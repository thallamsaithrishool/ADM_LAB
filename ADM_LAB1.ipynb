```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data=pd.read_csv('/content/mental_health_dataset.csv')
data.head()
```

```python
print("missing values before imputation:")
print(data.isnull().sum())
```

```python
data['Work_Hours'].fillna(data['Work_Hours'].mean(), inplace=True)
```

```python
data['Age'].fillna (data['Age'].mean(), inplace=True)
data['Work_Hours'].fillna(data['Work_Hours'].mode()[0], inplace=True)
```

```python
print("\nMissing values after imputation:")
print(data.isnull().sum())
```

```python
scaler_minmax = MinMaxScaler()
columns_to_normalize = ['Age', 'Work_Hours']
data[columns_to_normalize]= scaler_minmax.fit_transform(data[columns_to_normalize])
```

```python
data.head()
```

```python
scaler_standard = StandardScaler()
data[columns_to_normalize] = scaler_standard.fit_transform(data[columns_to_normalize])
```
