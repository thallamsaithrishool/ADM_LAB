```python
  import pandas as pd
  data=pd.read_csv('/content/tested.csv')
```

```python
summary=data.describe()
print(summary)
```

```python
print(data.describe(include=[object]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['Age'].dropna(),bins=20,edgecolor='black',color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['Pclass'].value_counts().plot(kind='bar',color='lightblue')
plt.title('Passenger Class Distribution')
plt.xlabel(' Class')
plt.ylabel('Count')
plt.xticks(rotation=1)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['Age'].dropna(),patch_artist=True,boxprops=dict(facecolor='black'))
plt.title('Box Plot Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['Age','Fare','SibSp','Parch']].dropna())
plt.show()
```

```python
correlation_matrix=data[['Age','Fare','SibSp','Parch']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(x='Survived',data=data,palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Pclass',y='Age',data=data,palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.ylabel('Age')
```
